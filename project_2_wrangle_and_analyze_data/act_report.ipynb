# Analysis Report
On Twitter, one of the more popular user accounts is [WeRateDogs](https://twitter.com/dog_rates), which at the time of writing this report had 9.2 million followers. The account, as per the user bio, is "your only source for professional dog ratings". The user posts several pictures of dogs in different, often humorous, contexts and provides ratings for the dogs out of 10. One unique aspect of this rating system is the numerator is almost always greater than 10, meaning each dog scores more than 100%. Twitter users frequently interact with this account's tweets, with a majority receiving hundreds, if not thousands of retweets, in addition to thousands of likes. One question however is, is there more to these interactions? Are there certain variables that influence the number of likes and retweet that a tweet gets? This report attempts to answer some of these questions, and analyzes tweets posted between November 2015 and August 2017.

The following is the list of questions that this report attempts to answer:
1. Among the images of dogs posted on the WeRateDogs Twitter account, which dog breed appears the most?
2. Which dog breeds received the highest average number of likes per tweet?
3. Of top 10 dog breeds that were most posted on the account, which ones were retweeted the most on average?
4. Which dog stage, according to the [Dogtionary](https://video.udacity-data.com/topher/2017/October/59e04ceb_dogtionary-combined/dogtionary-combined.png) received the highest number of likes?
5. What was the tweet frequency of the account user, during the duration this analysis was applied to?
6. How did the average like count change between November 2015 and August 2017?

### Most Frequently Posted Dog Breed
Out of 114 dog breeds sampled, the golden retriever was posted most, appearing in 156 tweets, followed by the Labrador Retriever with 103 tweets and the Pembroke with 94 tweets. The top 10 dog breeds were:
  - golden_retriever (156 tweets)
  - Labrador_retriever (103 tweets)
  - Pembroke (94 tweets)
  - Chihuahua (90 tweets)
  - pug (62 tweets)
  - toy_poodle (49 tweets)
  - chow (48 tweets)
  - Pomeranian (41 tweets)
  - Samoyed (41 tweets)
  - malamute (33 tweets)

### Dog breeds with the highest average number of tweets
The dog breeds with the highest average number of likes were:
   - Bedlington Terrier (6 tweets in total) - 21570
   - Saluki (4 tweets in total) - 20822 likes per tweet
   - French Bulldog (30 tweets in total) - 17554 likes per tweet

### Dog Stage that Received the Highest Average Number of Likes
The four stages considered were doggo, puppo, pupper and floofer. Some of the images had a doggo and a pupper together, hence the additional category. Out of these, the most liked on average was the puppo stage, which is the intermediate stage between a pupper and a doggo.
